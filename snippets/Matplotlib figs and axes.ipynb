```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(nrows=3, ncols=6)
axs[0,0].plot(np.arange(5), 2*np.arange(5))
axs[0,0].set_title('Tit 1')
axs[0,1].plot(np.arange(5), -3*np.arange(5))
axs[0,1].set_title('Tit 2')
axs[0,4].set_visible(False)
fig.suptitle('Big title')
plt.show()
```
